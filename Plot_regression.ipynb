# Normal vs MA

```python
import torch.nn.functional as F
# 假設你要載入的模型或者權重是存儲在檔案 normal_MA_x_img.pth

# 假設你要載入的模型或者權重是存儲在檔案 normal_MA_x_img.pth
model_path = '/home/jding/Documents/PD_contrastive_research_0817/Age_effect/Normal_MA_pred.pth'

# 使用 torch.load() 來載入該檔案
model_predict = torch.load(model_path)
prob = F.sigmoid(model_predict )[:,1].cpu().detach().numpy()
# 假設你要載入的模型或者權重是存儲在檔案 normal_MA_x_img.pth

# 假設你要載入的模型或者權重是存儲在檔案 normal_MA_x_img.pth
x_f_path = '/home/jding/Documents/PD_contrastive_research_0817/Age_effect/Normal_MA_x_f.pth'

# 使用 torch.load() 來載入該檔案
x_f = torch.load(x_f_path)

x_f_age = np.array(pd.DataFrame(x_f.cpu().detach().numpy()).loc[:,0])

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import linregress

# 假設 x 和 y 是長度為 206 的樣本點，這裡使用 x_f_age 和 prob 作為輸入資料
# np.random.seed(42)  # 固定隨機種子以重現結果
# x_f_age = np.random.rand(206) * 100  # 假設年齡分布
# prob = 2.5 * x_f_age + np.random.normal(0, 50, 206)  # 假設的結果

x_f_age_m = x_f_age
prob_m =prob

# 設定圖形風格為Nature風格
sns.set_theme(style="whitegrid")
sns.set_context("paper", font_scale=1.5)

# 計算線性回歸並獲取統計值
slope, intercept, r_value, p_value, std_err = linregress(x_f_age_m, prob_m)

# 計算 R^2
r_squared = r_value**2

# 設定figure大小
plt.figure(figsize=(6, 6))

# 繪製回歸圖
sns.regplot(x=x_f_age_m, y=prob_m, scatter_kws={'s': 80, 'color': 'black'}, line_kws={'color': 'red', 'lw': 2})

# 設定圖表標籤和標題樣式
plt.xlabel("Age", fontsize=16, weight='bold')
plt.ylabel("Predicted value", fontsize=16, weight='bold')
plt.title("Normal vs MA", fontsize=18, weight='bold')

# 設定圖形邊界線樣式
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_linewidth(1.5)
plt.gca().spines['left'].set_linewidth(1.5)

# 調整x, y 軸字體大小
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# 顯示 R^2 與 p-value 於圖表上，以科學記號表示 p-value
legend_text = f'R² = {r_squared:.3f}\np-value = {p_value:.2e}'
plt.legend([legend_text], loc='upper left', bbox_to_anchor=(0.04, 0.95), fontsize=14, frameon=False)

# 顯示圖形
plt.tight_layout()

plt.savefig("/home/jding/Music/PD_期刊文獻蒐集/PD_paper_Fig/Normal_MA_regression_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```
